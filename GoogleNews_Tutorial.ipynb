# Usage

## Setting

### 寫法1

```python
from GoogleNews import GoogleNews
googlenews = GoogleNews()

# 設置前先清除上次爬蟲的資料
googlenews.clear()

# 選擇語言
googlenews.set_lang('zh-TW') # e.g., en, fr, jp, zh(中文簡體)

# 設定發布時間(set_period 和 set_time_range 只能選一個使用)
googlenews.set_period('w') # h(過去一小時), d(過去一天), w(過去一周), m(過去一個月), y(過去一年)
# googlenews.set_time_range('06/21/2022','06/28/2022') # 月/日/年

# encoding
googlenews.set_encode('utf-8')
```

### 寫法2

```python
from GoogleNews import GoogleNews

# 設置前先清除上次爬蟲的資料
GoogleNews().clear()

# setting(period 和 start end 只能選一個使用)
googlenews = GoogleNews(lang='zh-TW', period='w', encode='utf-8')
# googlenews = GoogleNews(lang='zh-TW', start='06/21/2022', end='06/28/2022', encode='utf-8')
```

# Search

```python
# keyword
googlenews.search('蔡英文')
```

# Result
image.png

## 只要第一頁(前10筆)

### 寫法1

```python
result = googlenews.page_at(1)
print(len(result))
result
```

### 寫法2

```python
result = googlenews.results() # 依關聯性排序
# result = googlenews.results(sort = False) # 依發布時間排序(新->舊) (sort = True會有error，應該是api沒寫好)
print(len(result))
result
```

## 指定某一頁 (只能依關聯性排序)

```python
result = googlenews.page_at(3) # 指定第三頁的搜尋結果
print(len(result))
result
```

## 指定某頁到某頁 (只能依關聯性排序)

竟然沒有fuction可以call

```python
result = []
for i in range(1,4):  # 指定第一頁到第三頁
  googlenews.page_at(i)
  result.extend(googlenews.page_at(i))
  # print("i = ", i, "\n", googlenews.page_at(i))

print(len(result))
result
```
